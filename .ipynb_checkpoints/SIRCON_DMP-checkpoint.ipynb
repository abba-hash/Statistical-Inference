```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from scipy.integrate import solve_ivp

# Network parameters
n = 100  # Number of nodes
m = 2   # Number of edges to attach from a new node to existing nodes
G = nx.barabasi_albert_graph(n, m)

# Time span for the simulation
t_span = (0, 10)
t_eval = np.linspace(*t_span, 10)

# Parameters
beta = 0.1
gamma = 0.3

# Initial conditions
S0 = np.ones(n)  # All nodes start fully susceptible
I0 = np.zeros(n)
I0[0] = 0.5  # Initially infect the first node
S0[0] -= I0[0] 
R0 = np.zeros(n)

# Flatten initial conditions for ODE solver
y0 = np.concatenate([S0, I0, R0])


# SIR dynamics with rDMP approach
def SIRCON_DMP(t, y, G, beta, gamma):
    S, I, R = np.split(y, 3)
    dSdt = np.zeros(n)
    dIdt = np.zeros(n)
    dRdt = np.zeros(n)
    
    for node in G.nodes:
        S[node] -= beta * S[node] * sum(I[neighbor] for neighbor in G.neighbors(node))
        I[node] += beta * S[node] * sum(I[neighbor] for neighbor in G.neighbors(node)) - gamma * I[node]
        R[node] += gamma * I[node] 
        
    return np.concatenate([dSdt, dIdt, dRdt])

# Solve the SIR model with rDMP
solution = solve_ivp(SIRCON_DMP, t_span, y0, args=(G, beta, gamma), t_eval=t_eval, method='RK45')

# # Plotting
plt.figure(figsize=(12, 6))

# # Split the results
S, I, R = np.split(solution.y, 3)
node_index = 0
plt.plot(t_eval, I[node_index], label='Infected', color='orange')
plt.plot(t_eval, S[node_index], label='Susceptible', color='blue')
plt.plot(t_eval, R[node_index], label='Recovered', color='green')
plt.xlabel('Time')
plt.ylabel('Fraction of Population')
plt.legend()
plt.title(f'Disease Dynamics in Node {node_index}')
plt.show()
```
